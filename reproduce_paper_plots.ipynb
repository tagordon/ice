```python
import numpy as np
import matplotlib.pyplot as plt

import french
```

```python
rho = np.linspace(1, 4.5, 100)
plt.plot(rho, rho * rho * french.un(rho)[1])
```

```python
rho = np.linspace(0.1, 4.5, 100)
T = [295, 400, 500, 600, 800, 1000, 1250, 1500, 1750, 2000]

for T in T:
    s = french.ft(rho, T)[1] * 18000 / (6.022 * 1.38 * 3)
    plt.plot(rho, s)
    
plt.xlim(0, 4.5)
plt.ylim(0, 4)
```

```python
rho = np.linspace(1.5, 4.5, 100)

models = ['PBE', 'HSE', 'vdW', 'AM05', 'LDA']

fig, axs = plt.subplots(1, 2, figsize=(12, 5))

for m in models:
    axs[0].plot(rho, french.ue(rho, model=m)[0], label=m)
    
for m in models:
    axs[1].plot(rho, rho * french.ue(rho, model=m)[1], label=m)
    
plt.legend()
```

```python
rho = np.linspace(1.5, 3.75, 1000)

for m in models:
    p_reduced = french.french(rho, 295, model=m)[2] / rho
    plt.plot(rho, p_reduced)
    
plt.xlim(1.5, 3.75)
```
